### Statistics and analysing the correlations in the given dataset

```python
correlation = data_set.corr()
correlation
```

## Data Visualization in the dataset

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data_set)
plt.title('Category V/S Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=90)
plt.show()
```

```python
data_set.hist(bins=60,figsize=(20,15))
plt.show()
```

```python
print(data_set['State'].value_counts())
plt.figure(figsize=(15,10))
sns.countplot(x=data_set['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Category',data=data_set)
plt.title('DISTRIBUTION OF CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.subplot(2,1,2)
sns.countplot('Sub-Category',data=data_set)
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.show()

```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Segment',data=data_set)
plt.grid(alpha=0.3)
plt.subplot(2,1,2)
sns.countplot('Region',data=data_set)
plt.grid(alpha=0.3)
plt.show()

```

## Heatmap of Correlation

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```
